```python
import time
import random
import csv
from bs4 import BeautifulSoup
import requests

url_list = []
for i in range(0, 600, 30):
    url_list.append('https://www.yelp.com/search?find_loc=New+York,+NY&start='+str(i)+'&cflt=restaurants')
#print(url_list)

def scrape_reviews(reviews, csvwriter):
    for review in reviews:
        dic = {}
        username = review.find('a', attrs={'class': 'biz-name js-analytics-click'}).text
        #location = review.find('span', attrs={'class': 'neighborhood-str-list'}).text.strip()
        rating = review.find('img', attrs={'class': 'offscreen'}).get('alt')
        review_count = review.find('span', attrs={'class': 'review-count rating-qualifier'}).text
        cuisine = review.find_all('span', attrs={'class': 'category-str-list'})
        cuisine = ', '.join([c.getText() for c in cuisine if c.getText()])
        #cuisine.strip('\n').replace('\"','')
        #review_count.strip('\n').replace('\"','')
        #Price = review.find('span', attrs={'class': 'business-attribute price-range'}).text
      
        try:
            Price = review.find('span', attrs={'class': 'business-attribute price-range'}).text
        except:
            Price = ""
            
        try:
            location = review.find('span', attrs={'class': 'neighborhood-str-list'}).text.strip()
        except:
            location = ""
            
        dic['username'] = username
        dic['location'] = location
        dic['rating'] = rating
        dic['review_count'] = review_count
        dic['cuisine'] = cuisine
        dic['Price'] = Price
        
        review_writer.writerow(dic.values())
    

with open('all_reviews.csv', 'w') as csvfile:
    review_writer = csv.writer(csvfile)
    for index, url in enumerate(url_list):
        response = requests.get(url, headers=headers).text
        soup = BeautifulSoup(response, 'html.parser')
        reviews = soup.find_all('div', attrs={'class':'search-result natural-search-result scrollable-photos-search-result'})
        scrape_reviews(reviews, review_writer)
        # Random sleep to avoid getting banned from the server
        time.sleep(random.randint(1,3))
        # Log the progress
        print('Finished page ' + str(index + 1))
```
